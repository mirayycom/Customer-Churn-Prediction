### Feature Engineering

```python
# scikit encoder kutuphanesini kullanabilmek için import ettim


from sklearn.preprocessing import LabelEncoder
```

```python
# Status alanını numeric yapabilmek icin etiketleme yonetimini kullandım 

data['Status'] = LabelEncoder().fit_transform(data['Status'])
data.head()
```

```python
# böylece Status alanı 0-1 degerlerinden olusan, integer veri tipine donustu

data.groupby('Status')['Status'].count()
```

```python
# Total_Trans_Amt değişkeni için aykırı gözlem analizinin yapılması ve aykırı değerlerin çıkarılması

q1 = data["Total_Trans_Amt"].quantile(0.25)
q3 = data["Total_Trans_Amt"].quantile(0.75)

IOC = q3 - q1

alt_sınır = q1 - 1.5*IOC
üst_sınır = q3 + 1.5*IOC

kontrol = (data["Total_Trans_Amt"] < alt_sınır) | (data["Total_Trans_Amt"] > üst_sınır)
data["overvalue"] = kontrol
data["overvalue"].value_counts()
```

```python
# Fiyat değişkeninde aşırı değer olarak 759 tane satır gözüküyor bu değerleri çıkaralım.

data = data.loc[data["overvalue"] == False]
data.head()
```

```python
# Kart Kategorilerinin Ortalama İşlem Hacmi

round(pd.pivot_table(data = data, columns = "Card_Category", values = "Total_Trans_Amt")).T

# buradan anladığım encoding ile kategorileri önem derecesine göre numaralandırabilirim
```

```python
def split_and_test(X, y):
    '''
    For a set of features and target X, y, perform a 60/40 train/test split, 
    fit and test a linear regression model, and report results
    '''
    
    # perform train/test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
    
    # fit linear regression to training data
    lr_model = LinearRegression()
    lr_model.fit(X_train, y_train)
    
    # score fit model on test data
    test_score = lr_model.score(X_test, y_test)
    
    # report results
    print('Test R^2 score was:', test_score)
```
